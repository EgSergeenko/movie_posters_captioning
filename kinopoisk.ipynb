## Поиск фильмов на Кинопоиске

Требуется оплатить 500р 

```python
kp_reponses = {}
```

```python
for row in tqdm(list(kp.iterrows())[1000:]):
    row_id = row[0]
    text = str(row[1]['NAME']) + ' ' + str(row[1]['YEAR'])    
    
    if row_id not in kp_reponses.keys():
        print(row_id, text)
        
        url = f'https://kinopoiskapiunofficial.tech/api/v2.1/films/search-by-keyword?keyword={text}'
        headers = {
            'X-API-KEY': 'd0105144-3479-4987-8e61-6e3538fd2d6f',
            'Content-Type': 'application/json',
        }
        r = requests.get(url, headers=headers)
        response = r.json()
        
        if 'status' in response.keys():
            break
        
        kp_reponses[row_id] = [text, response]
```

## Парсинг ответов

```python
films = {}

for i, v in tqdm(kp_reponses.items()):
    for f in v[1]['films']:
        if 'description' in f.keys():
            films[f['filmId']] = {
                'nameRu': f.get('nameRu', np.NaN),
                'nameEn': f.get('nameEn', np.NaN),
                'type': f['type'],
                'year': f['year'],
                'description': f['description'],
                'countries': f['countries'],
                'genres': f['genres'],
                'rating': f['rating'],
                'posterUrl': f['posterUrl']
            }
            
films = pd.DataFrame(films).T
```

## Выгрузка изображений

```python
for i, f in tqdm(films.iterrows()):
    urllib.request.urlretrieve(f['posterUrl'], f"data/img/{i}.jpg")
```
